# Shell Game

```python
from random import shuffle

def shuffle_list(mylist):
    shuffle(mylist)
    return mylist

def player_guess():
    guess=''
    while guess not in ['0','1','2']:
        guess = input('Pick a number: 0, 1 or 2   ') 
    return int(guess)

def check_guess(mylist,guess):
    if mylist[guess] == 'O':
        print('Correct!')
        print(mylist)
    else:
        print('Wrong guess!')
        print(mylist)
        
# FINAL SCRIPT:

mylist = [' ','O',' ']
mixedup_list = shuffle_list(mylist)
guess = player_guess()
check_guess(mixedup_list,guess)
```

# Rock paper scissors

```python
from random import shuffle

options = ['R','P','S']

def computer():
    shuffle(options)
    return options[0]

def player():
    player_choice = ''
    while True:
        player_choice = input('Choose rock, paper or scissors (enter R, P or S):').upper()
        if player_choice in options:
            return player_choice
        else:
            print('Wrong input, enter R, P or S')

def comparison(computer_choice, player_choice):
    if computer_choice == player_choice:
        print('Draw! Try again')
        game()
    elif computer_choice == 'R' and player_choice == 'P':
        print('Paper beats rock \nPlayer wins')
        return True
    elif computer_choice == 'P' and player_choice == 'R':
        print('Paper beats rock \nComputer wins')
        return False
    elif computer_choice == 'R' and player_choice == 'S':
        print('Rock beats scissors \nComputer wins')
        return False
    elif computer_choice == 'S' and player_choice == 'R':
        print('Rock beats scissors \nPlayer wins')
        return True
    elif computer_choice == 'P' and player_choice == 'S':
        print('Scissors beats paper \nPlayer wins')
        return True
    elif computer_choice == 'S' and player_choice == 'P':
        print('Scissors beats paper \nComputer wins')
        return False

def game():
    computer_choice = computer()
    player_choice = player()
    print(f'Computer chose {computer_choice}, Player chose {player_choice}')
    comparison(computer_choice, player_choice)
    
game()
```

# TIC TAC TOE

```python
from IPython.display import clear_output

def display(row_A,row_B,row_C):
    clear_output()
    print('     0    1    2')
    print('A ', row_A)
    print('B ', row_B)
    print('C ', row_C)
    
def player_row(PlayerNo):            
    guess_row=''
    while guess_row not in ['A','B','C']:
        guess_row = input(f'Player {PlayerNo} - row A, B or C?').upper()
        if guess_row not in ['A','B','C']:
            print('Wrong input, choose A, B or C')
    
    return guess_row

def player_column(PlayerNo):
    guess_column=''
    while guess_column not in ['0','1','2']:
        guess_column = input(f'Player {PlayerNo} - column 0, 1 or 2?')
        if guess_column not in ['0','1','2']:
            print('Wrong input, choose 0, 1 or 2')
    return int(guess_column)

def change(guess_row,guess_column, row_A, row_B, row_C,sign):
    rows={'A':row_A,'B':row_B,'C':row_C}
    if rows[guess_row][guess_column]==' ':
        rows[guess_row][guess_column]=sign
        return False
    else:
        display(row_A,row_B,row_C)
        print('This possition is already taken, try again!')
        return True
    
def check_win(row_A,row_B,row_C):
    result = "Good Game!"
    check = True
    
    results = ["Player 1 has won!", "Player 2 has won!"]
        
    if row_A[0] == row_A[1] == row_A[2] != ' ':
        check = False
        result = results[0 if row_A[0] == 'X' else 1]
    elif row_B[0] == row_B[1] == row_B[2] != ' ':
        check = False
        result = results[0 if row_B[0] == 'X' else 1]
    elif row_C[0] == row_C[1] == row_C[2] != ' ':
        check = False
        result = results[0 if row_C[0] == 'X' else 1]
    elif row_A[0] == row_B[0] == row_C[0] != ' ':
        check = False
        result = results[0 if row_A[0] == 'X' else 1] 
    elif row_A[1] == row_B[1] == row_C[1] != ' ':
        check = False
        result = results[0 if row_A[1] == 'X' else 1] 
    elif row_A[2] == row_B[2] == row_C[2] != ' ':
        check = False
        result = results[0 if row_A[2] == 'X' else 1] 
    elif row_A[2] == row_B[2] == row_C[2] != ' ':
        check = False
        result = results[0 if row_A[2] == 'X' else 1] 
    elif row_A[0] == row_B[1] == row_C[2] != ' ':
        check = False
        result = results[0 if row_A[0] == 'X' else 1]
    elif row_A[2] == row_B[1] == row_C[0] != ' ':
        check = False
        result = results[0 if row_A[2] == 'X' else 1]
    else:
        pass
    return check, result

def tictactoe():
    row_0=['0','1','2']
    row_A=[' ',' ',' ']
    row_B=[' ',' ',' ']
    row_C=[' ',' ',' ']

    signs=['X','O']
    players=[1,2]
    
    check=True
    result="Good Game!"

    for x in range(0,9):
        player_index=x%2
        repeat=True
   
        while repeat:
            guess_row = player_row(players[player_index])
            guess_column = player_column(players[player_index])
            repeat = change(guess_row,guess_column, row_A, row_B, row_C,signs[player_index])
            
        display(row_A,row_B,row_C)
        
        check, result = check_win(row_A,row_B,row_C)
        
        if not check:
            break
    return result

tictactoe()
```

# WAR

```python
import random 

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':11,
         'Queen':12, 'King':13, 'Ace':14}

class Card:
    
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
    
    def __str__(self):
        return self.rank + " of " + self.suit

    
class Deck:
    
    def __init__(self):
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                # create Card object
                created_card=Card(suit,rank)
                self.all_cards.append(created_card)
    def shuffle(self):
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        return self.all_cards.pop()

class Player:
    
    def __init__(self,name):
        self.name = name
        self.all_cards = []
        
    def remove_one(self):
        return self.all_cards.pop(0)
    
    def last_card(self):
        return self.all_cards.pop
    
    def add_cards(self,new_cards):
        if type(new_cards) == type ([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
            
    def __str__(self):
        return f'Player {self.name} has {len(self.all_cards)} cards.'

# Game logic:

Player1 = Player("Player One")
Player2 = Player("Player Two")
deck=Deck()
deck.shuffle()

for i in range(0,52,2):
    Player1.add_cards(deck.deal_one())
    Player2.add_cards(deck.deal_one())
print(Player1)
print(Player2) 

game_on = True

while game_on:
    card1=Player1.remove_one()
    card2=Player2.remove_one()
    new_cards=[card1,card2]
    print(f'Player 1 has {len(Player1.all_cards)} cards, Player 2 has {len(Player2.all_cards)} cards')
    print(f'Player 1 has {card1}, Player 2 has {card2}')
    if card1.value > card2.value:
        Player1.add_cards(new_cards)
        print(f'Player 1 takes the cards.')
    elif card1.value < card2.value:
        Player2.add_cards(new_cards)
        print(f'Player 2 takes the cards.')
    elif card1.value == card2.value:
        print("WAR!")
        at_war = True
        war_cards=[card1, card2]
        
        while at_war:
            if len(Player1.all_cards) > 2 and len(Player2.all_cards) > 2:
                table_cards=[Player1.remove_one(),Player1.remove_one(),Player2.remove_one(),Player2.remove_one()]
                card1=Player1.remove_one()
                card2=Player2.remove_one()
                print(f'Player 1 has {card1}, Player 2 has {card2}')
                if card1.value > card2.value:
                    Player1.add_cards(war_cards)
                    Player1.add_cards(card1)
                    Player1.add_cards(card2)
                    Player1.add_cards(table_cards)
                    print("Player 1 wins the war!")
                    at_war = False
                    break
                elif card1.value < card2.value:
                    Player2.add_cards(war_cards)
                    Player2.add_cards(card1)
                    Player2.add_cards(card2)
                    Player2.add_cards(table_cards)
                    print("Player 2 wins the war!")
                    at_war = False
                    break
                elif card1.value == card2.value:
                    war_cards.extend(table_cards)
                    war_cards.append(card1)
                    war_cards.append(card2)
                    print('War continues...')
            elif len(Player1.all_cards) < 3:
                print("Player 1 has not enough cards for war, Player 2 wins!")
                game_on = False
                at_war = False
                break
            elif len(Player2.all_cards) < 3:
                print("Player 2 has not enough cards for war, Player 1 wins!")
                game_on = False
                at_war = False
                break
        
    if len(Player1.all_cards) == 0:
        print("Player 2 wins!")
        game_on = False
        break
    if len(Player2.all_cards) == 0:
        print("Player 1 wins!")
        game_on = False
        break
```
