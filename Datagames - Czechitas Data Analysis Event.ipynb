# Purchase

In the attached file you have a list where each line represents the types of goods purchased by one customer. Find the pair of foods that is most often found in them.

```python
df = pd.read_csv('validation_input_grocery.csv', sep =';', header=None, names=range(21))
df.head()
```

```python
import csv

with open('validation_input_grocery.csv') as f:
    reader = csv.reader(f, delimiter=';')
    df_list = list(reader)

df_list[0]
```

```python
from itertools import combinations
from collections import Counter

def most_freq(l, n):
    temp=[]
    for i in l:
        temp.extend(tuple(k) for k in combinations(i, n))
    res = Counter(temp)
    return res
```

```python
result = most_freq(df_list,2)
result.most_common()[0]
```

# Cards

Calculate the results of the tournament.

The card game is a bit like poker. Each player is dealt 6 cards and each card is marked with two letters - a suit (H, T, C, P - represent clubs, diamonds, hearts and spades) and a value (numbers from 2 to 9 and the letters J (Jack), Q (Queen), K (King) and A).

The game has the following rules:

Each player has 6 cards.
The player who has the most valuable combination of cards wins. He gets 1 point. If two players have the same value combination, they both get a point.
The combinations ranked from least valuable to most valuable are:
- two cards of the same value,
- three cards of the same value,
- sequence of five consecutive cards (regardless of suit)
- four cards of the same value.
If none of the players get a scoring combination, no one gets a point.

To determine the winner, it depends only on what combination the player has. For example, if one player has two 3 cards and the other player has two Q cards, they both score a point.

Players Alice, Bob, Claire, Dee, and Elizabeth participated in the tournament. Find out which player had the most points at the end and how many they had. Write the result in the format Name Points, e.g. Claire 4.

In the data file you always have the cards that each player had in their hand at the end of the game. The individual cards are separated by a dash and the players are separated by a semicolon, the order of the players is alphabetical.

The order for the card sequence is 23456789JQKA (ie A is recognized only at the very end, not at the beginning).

```python
df = pd.read_csv('validation_input_cards.csv', sep =';', index_col = False, names=['Alice', 'Bob', 'Claire', 'Dee', 'Elizabeth'])
df.head()
```

```python
def calc_points(hand):
    mylist = hand.split('-')
    cards=[]
    for i in mylist:
        cards.append(i[1])
    y=0
    for x in cards:
        if cards.count(x) > y:
            y=cards.count(x)
    if y == 2:
        return 1
    if y == 3:
        return 2
    if y == 4:
        return 4
    numbers = []
    for x in cards:
        if x == 'J':
            y = 10
        elif x == 'Q':
            y = 11
        elif x == 'K':
            y = 12
        elif x == 'A':
            y = 13
        else:
            y = int(x)
        numbers.append(y)
    
    numbers = sorted(numbers)
    z=0
    for x in range(1,6):
        if numbers[x] == numbers[x-1] + 1:
            z += 1  
    if z >= 4:
        return 3
    else:
        return 0
```

```python
for x in df.columns:
    df[x] = df[x].apply(calc_points)
    
df
```

```python
for x in df.columns:
    name = x
    score = df[x].sum()
    print(f'{name}: {score}')
```
