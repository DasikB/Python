```python
import csv

with open('validation_input_grocery.csv') as f:
    reader = csv.reader(f, delimiter=';')
    df_list = list(reader)

df_list
```

```python
from itertools import combinations
from collections import Counter

def most_freq(l, n):
    temp=[]
    for i in l:
        temp.extend(tuple(k) for k in combinations(i, n))
    res = Counter(temp)
    return res
```

```python
tabulka = most_freq(df_list,2)
tabulka.most_common()[0]
```

# Karty

Dopočítej výsledky turnaje.

Karetní hra je trochu podobná pokeru. Každý hráč dostane 6 karet a každá karta je označena dvěma písmeny - barvou (H T, C, P - označují kříže, káry, srdce a píky) a hodnotou (čísla od 2 do 9 a písmeny J (Jack), Q (Queen), K (King) a A).

Hra má následující pravidla:

Každý hráč má 6 karet.
Vyhrává ten hráč, který má v kartách tu nejhodnotnější kombinace. Ten dostane 1 bod. Pokud mají dva hráč stejně hodnotnou kombinaci, získávají bod oba.
Kombinace seřazené od nejméně hodnotné po nejhodnotnější jsou:

    - dvě karty stejné hodnoty,
    - tři karty stejné hodnoty,
    - posloupnost pěti po sobě jdoucích karet (bez ohledu na barvu)
    - čtveřice karet stejné hodnoty.
Pokud nikdo z hráčů nezíská žádnou bodovanou kombinaci, nezískává bod nikdo.

U určení vítěze závisí pouze na tom, jakou kombinaci hráč má. Pokud například má jeden hráč dvě karty hodnoty 3 a druhý hráč má dvě karty hodnoty Q, získávají bod oba.

Turnaje se zůčastnili hráči Alice, Bob, Claire, Dee, a Elizabeth. Zjisti, který hráč měl na konci nejvíce bodů a kolik jich měl. Výsledek zapiš ve formátu Jméno Body, např. Claire 4.

V datovém souboru máš vždy karty, které měl na konci hry každý hráč v ruce. Jednotlivé karty jsou oddělené pomlčkou a hráči jsou odděleni středníkem, pořadí hráčů je dle abecedy.

Pořadí pro posloupnost karet je 23456789JQKA (tj. A se uznává pouze na úplném konci, nikoli na začátku).

```python
df = pd.read_csv('validation_input_cards.csv', sep =';', index_col = False, names=['Alice', 'Bob', 'Claire', 'Dee', 'Elizabeth'])
df.head()
```

```python
def cards(vstup):
    mylist = vstup.split('-')
    karty=[]
    for i in mylist:
        karty.append(i[1])
    y=0
    for x in karty:
        if karty.count(x) > y:
            y=karty.count(x)
    if y == 2:
        return 1
    if y == 3:
        return 2
    if y == 4:
        return 4
    cisla = []
    for x in karty:
        if x == 'J':
            y = 10
        elif x == 'Q':
            y = 11
        elif x == 'K':
            y = 12
        elif x == 'A':
            y = 13
        else:
            y = int(x)
        cisla.append(y)
    
    cisla = sorted(cisla)
    z=0
    for x in range(1,6):
        if cisla[x] == cisla[x-1] + 1:
            z += 1  
    if z >= 4:
        return 3
    else:
        return 0
```

```python
df["Alice"] = df["Alice"].apply(cards)
df["Bob"] = df["Bob"].apply(cards)
df["Claire"] = df["Claire"].apply(cards)
df["Dee"] = df["Dee"].apply(cards)
df["Elizabeth"] = df["Elizabeth"].apply(cards)
df
```

```python
Alice = df.Alice.sum()
print(f"Alice: {Alice}")
Bob = df.Bob.sum()
print(f"Bob: {Bob}")
Claire = df.Claire.sum()
print(f"Claire: {Claire}")
Dee = df.Dee.sum()
print(f"Dee: {Dee}")
Elizabeth = df.Elizabeth.sum()
print(f"Elizabeth: {Elizabeth}")
```
